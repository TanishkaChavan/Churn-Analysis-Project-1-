```python
Project Description:
The customer churn dataset is a collection of customer data that focuses on predicting customer churn, which refers to the tendency of
customers to stop using a companys products or services. The dataset contains various features that describe each customer, such as their 
credit score, country, gender, age, tenure, balance, number of products, credit card status, active membership, estimated salary, and churn status. 
The churn status indicates whether a customer has churned or not. The dataset is used to analyze and understand factors that contribute to customer
churn and to build predictive models to identify customers at risk of churning. The goal is to develop strategies and interventions to reduce churn 
and improve customer retention.
```

```python
Data Description:
    The dataset contains 12 columns and 10,000 rows.
    Key features include:
      - Numerical columns: credit_score, age, tenure, balance, estimated_salary, etc.
      - Categorical columns: country, gender.
      - Target variable: churn (indicating if a customer has churned).
      -	All columns have 10,000 non-null entries.
```

```python
# Age distribution
sns.histplot(df['age'], kde=True, bins=30, color='red')
plt.title("Age Distribution")
plt.show()
```

```python
# Balance distribution
sns.histplot(df['balance'], kde=True, color='green')
plt.title("Balance Distribution")
plt.show()
```

```python
# Churn count
sns.countplot(x='churn', data=df, palette='Set2')
plt.title("Churn Count")
plt.xticks([0, 1], labels=['Not Churned', 'Churned'])
plt.show()

# Churn percentage
churn_rate = df['churn'].value_counts(normalize=True) * 100
print(churn_rate)
```

```python
# What is the impact of age and tenure on churn?
# Age vs Churn
sns.boxplot(x='churn', y='age', data=df, palette='Set3')
plt.title("Age vs. Churn")
plt.xticks([0, 1], labels=['Not Churned', 'Churned'])
plt.show()
```

```python
# Churn by Country
country_churn = pd.crosstab(df['country'], df['churn'], normalize='index') * 100
country_churn.plot(kind='bar', stacked=True, color=['pink', 'blue'], figsize=(8, 5))
plt.title("Churn by Country")
plt.xlabel("Country")
plt.ylabel("Percentage")
plt.legend(['Not Churned', 'Churned'])
plt.show()

# Churn by Gender
sns.countplot(x='gender', hue='churn', data=df, palette='pastel')
plt.title("Churn by Gender")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import math

# List of numerical columns (example columns)
numerical_column = ['credit_score', 'age', 'tenure', 'balance', 'products_number', 'estimated_salary']

# Calculate rows and columns for the grid
num_plots = len(numerical_column)
rows = math.ceil(num_plots / 3)  # Dynamic calculation
cols = 3

# Set the figure size
plt.figure(figsize=(15, rows * 4))

# Loop through numerical columns and create box plots
for idx, col in enumerate(numerical_column):
    plt.subplot(rows, cols, idx + 1)  # Dynamically set grid position
    sns.boxplot(x='churn', y=col, data=df, palette='coolwarm', hue=None)  # Set `hue=None`
    plt.title(f'Boxplot of {col} by Churn')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
numerical_df = df.select_dtypes(include=['float64', 'int64'])
# correlation matrix
corr_matrix = numerical_df.corr()
```

```python
# heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
